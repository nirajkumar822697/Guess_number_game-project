```python
import random
```

```python
computer_guess =random.randint(1,50)
```

```python
i=1
while(i<=5):
    num = int(input(f"Guess a number {i}: "))
    if num>=1 and num<=50:
        
        if num == computer_guess:
            print("Congrates you win 7cr")
            break
        elif num<computer_guess:
            print("Please guess a big number")
        elif num >computer_guess:
            print("please guess a small number")
    else:
        print("Think in limit 1 to 50")
        continue
        
    i+=1
else:
    print("You are looser!")
```
